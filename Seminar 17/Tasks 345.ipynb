## Task 3

```python
df = pd.read_csv("../Bonus19/BRCA_pam50.tsv", sep="\t", index_col=0)
X = df.iloc[:, :-1].to_numpy()
y = df["Subtype"].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=17
)
```

```python
model = RandomForestClassifier(
    class_weight="balanced",
    random_state=17
)
params = {'n_estimators': [10, 50, 200], 'max_depth': [3, 9]}
```

```python
clf = GridSearchCV(model, params, scoring=make_scorer(balanced_accuracy_score), 
                   n_jobs=-1, cv=RepeatedStratifiedKFold(n_repeats=10))
clf.fit(X, y)
```

```python
print(clf.best_params_)
y_pred = clf.predict(X_test)
round(balanced_accuracy_score(y_pred, y_test), 3)
```

## Task 4

```python
df = pd.read_pickle("bc_data.pkl")
ann = pd.read_pickle("bc_ann.pkl")
with open('genes_SVM.txt') as g:
    genes = g.read().replace("\n", "").split(";")
df_11 = df[genes]
```

```python
X_train = df.loc[ann.loc[ann["Dataset type"] == "Training"].index].to_numpy()
X_train_11 = df_11.loc[ann.loc[ann["Dataset type"] == "Training"].index].to_numpy()
y_train = ann.loc[ann["Dataset type"] == "Training", "Class"].to_numpy()
X_test = df.loc[ann.loc[ann["Dataset type"] == "Validation"].index].to_numpy()
X_test_11 = df_11.loc[ann.loc[ann["Dataset type"] == "Validation"].index].to_numpy()
y_test = ann.loc[ann["Dataset type"] == "Validation", "Class"].to_numpy()
```

```python
svm = SVC(kernel="linear", C=3.5)
svm.fit(X_train, y_train); pass
y_pred = svm.predict(X_test)
print("Balanced accuracy score:", round(balanced_accuracy_score(y_pred, y_test), 3))
M = confusion_matrix(y_test, y_pred)
print(M)
TPR = M[0, 0] / (M[0, 0] + M[0, 1])
TNR = M[1, 1] / (M[1, 0] + M[1, 1])
print("TPR:", round(TPR, 3), "TNR:", round(TNR, 3))
plot_roc_curve(svm, X_test, y_test)
plt.plot(1 - TPR, TNR, "x", c="red")
plt.show()
```

```python
svm = SVC(kernel="linear", C=3.5)
svm.fit(X_train_11, y_train); pass
y_pred = svm.predict(X_test_11)
print("Balanced accuracy score:", round(balanced_accuracy_score(y_pred, y_test), 3))
M = confusion_matrix(y_test, y_pred)
print(M)
TPR = M[0, 0] / (M[0, 0] + M[0, 1])
TNR = M[1, 1] / (M[1, 0] + M[1, 1])
print("TPR:", round(TPR, 3), "TNR:", round(TNR, 3))
plot_roc_curve(svm, X_test_11, y_test)
plt.plot(1 - TPR, TNR, "x", c="red")
plt.show()
```
